```python
X1=[0,0,1,1]
X2=[0,1,0,1]
op=[0,1,1,0]
w1=0.45
w2=0.56
b=0.33
number_of_epochs=1500
```

```python
import numpy as np 

def sigmoid(z):
    return 1/(1+np.exp(-z))

for epoch in range(1,number_of_epochs+1): 
    print(f'Epoch number: {epoch}')
    for iter in range(len(op)):
        z=w1*X1[iter]+w2*X2[iter]+b
        print(f'calculated z: {z}')
        P=sigmoid(z)
        print(f'Predicted Value: {P}')
        cost_val=(P-op[iter])**2
        print(f'cost value: {cost_val}')
        # print(f'iter number: {iter}')
    
    print('--'*25)

```

```python
import numpy as np 

h=0.0001
alpha=0.01
def cost(w1,w2,b,A,x1,x2):
    z=w1*x1+w2*x2+b
    p=1/(1+np.exp(-z))
    return (p-A)**2

for epoch in range(1,number_of_epochs+10000): 
    print(f'Epoch number: {epoch}')
    for iter in range(len(op)):
        d_J_w1=cost(w1,w2,b,op[iter],X1[iter],X2[iter])
        
        d_J_w1_h=cost(w1+h,w2,b,op[iter],X1[iter],X2[iter])
        d_w1=(d_J_w1_h-d_J_w1)/h

        d_J_w2_h=cost(w1,w2+h,b,op[iter],X1[iter],X2[iter])
        d_w2=(d_J_w2_h-d_J_w1)/h

        d_J_b_h=cost(w1,w2,b+h,op[iter],X1[iter],X2[iter])
        d_b=(d_J_b_h-d_J_w1)/h

        w1=w1-alpha*d_w1
        w2=w2-alpha*d_w2
        b=b-alpha*d_b

    print(w1,w2,b)
    print('--'*25)

```

```python
sigmoid(w1*1+w2*1+b)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers,models

model=models.Sequential()
model.add(layers.Dense(64,activation='relu',input_dim=2))
# model.add(layers.Dense(128,activation='relu'))
# model.add(layers.Dense(128*2,activation='relu'))
model.add(layers.Dense(1,activation='sigmoid'))

model.compile(loss='binary_crossentropy',optimizer='sgd',metrics=['accuracy'])
```

```python
import pandas as pd

data=pd.DataFrame(columns=['X1','X2','op'])
data['X1']=X1
data['X2']=X2
data['op']=op

X = data.drop(columns='op').values
y = np.array(data['op']).reshape(-1, 1)

model.fit(X,y,epochs=300)
```
